# Importing 
Checklist for all the Journal Citation Reports (JCR) indexed journals (20,0000+ in number) 

```python
data = pd.read_excel("latestJCRlist2022.xlsx")
```

# Creating a dataframe that contained only the columns of interest

```python
smalldf = data[['journal_name', 'if_2022','issn']]
```

# Filtering for Impact factor
Adding filters so as to obtain only those journals that had an impact factor between 2 and 5 (This was our target impact factor)

```python
targetif = smalldf.where((smalldf['if_2022'] > 2) & (smalldf['if_2022']<5)).dropna()
```

# RegEx for journal names
Using Regular Expression (RegEx) for the purpose of selecting Neurology/Neurosurgery related journals

```python
list1 = []

for journal in targetif['journal_name']:
    list1.append(list((re.findall("([\w ]*[N][e][u][r]+[\w ]*)(?:)", str(journal)))))
    list1.append(list((re.findall("([\w ]*[N][E][U][R]+[\w ]*)(?:)", str(journal)))))
list1
```

# Cleaning the List
Removing all the entries from the resultant list that were empty and hence of no use to us in the final dataset

```python
cleanlist = [ele for ele in list1 if ele != []]
```

# Selecting "journal_name" column as a separate dataset

```python
smalldf.index = smalldf['journal_name']
```

```python
smalldf
```

# Separate list
Created to obtain only the journals of interest

```python
for goal in cleanlist:
    print(goal)
```

```python
cleandf = pd.DataFrame(cleanlist)
```

```python
cleandf.index = cleandf[0]
```

```python
cleandf.drop(0, axis=1, inplace=True)
```

```python
cleandf
```

# Merging
The names of journals and the impact factor (if_2022) are merged together as a final dataset. The issn column is also retained

```python
finaldf = pd.merge(cleandf, smalldf, how = 'inner', left_index=True, right_index=True)
```

```python
finaldf.drop(['journal_name',1], axis = 1, inplace = True)
```

```python
finaldf.to_excel("Target NeuroJournals.xlsx")
```
